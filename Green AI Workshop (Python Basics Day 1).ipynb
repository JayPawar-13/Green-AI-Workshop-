```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operation = input("Enter the operation (+, -, *, /): ")

if operation == '+':
  result = num1 + num2
elif operation == '-':
  result = num1 - num2
elif operation == '*':
  result = num1 * num2
elif operation == '/':
  if num2 == 0:
    result = "Division by zero error"
  else:
    result = num1 / num2
else:
  result = "Invalid operation"

print(f"The result is: {result}")
```

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]
high_temp_threshold = 26
high_temp_cities = [city for city in climate_data if city["temperature"] > high_temp_threshold]
print("Cities with high temperature:")
for city in high_temp_cities:
    print(f"{city['city']} - {city['temperature']}°C")
```

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]
def avg():
  total_carbon = 0
  for city in climate_data:
    total_carbon += city["carbon_footprint"]
  average_carbon = total_carbon / len(climate_data)
  return average_carbon
avg()
```

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]
highets_footprint_city = max(climate_data, key=lambda city: city["carbon_footprint"])
print("City with the highest carbon footprint:")
print(f"{highets_footprint_city['city']} - {highets_footprint_city['carbon_footprint']} tonnes CO2")
```
